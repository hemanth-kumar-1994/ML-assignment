```python
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt

X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)
gmm = GaussianMixture(n_components=4, covariance_type='full', random_state=42)
gmm.fit(X)
y_gmm = gmm.predict(X)

plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_gmm, s=40, cmap='viridis')
plt.title('GMM Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Generating synthetic data for comparison
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)

# GMM clustering
gmm = GaussianMixture(n_components=4, covariance_type='full', random_state=42)
gmm.fit(X)
y_gmm = gmm.predict(X)

# K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Plotting the comparison between GMM and K-Means clustering
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# GMM Plot
ax1.scatter(X[:, 0], X[:, 1], c=y_gmm, s=40, cmap='viridis')
ax1.set_title("Gaussian Mixture Model (Soft Clustering)")
ax1.set_xlabel("Feature 1")
ax1.set_ylabel("Feature 2")

# K-Means Plot
ax2.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=40, cmap='viridis')
ax2.set_title("K-Means (Hard Clustering)")
ax2.set_xlabel("Feature 1")
ax2.set_ylabel("Feature 2")

plt.tight_layout()
plt.show()
```
